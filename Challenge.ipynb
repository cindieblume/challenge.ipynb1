#1. Análisis de facturación#


```python
# Agrego una columna 'id_tienda' que identifica cada tienda
tienda_1['id_tienda1'] = 'Tienda 1'
tienda_2['id_tienda2'] = 'Tienda 2'
tienda_3['id_tienda3'] = 'Tienda 3'
tienda_4['id_tienda4'] = 'Tienda 4'
```

```python
import pandas as pd

# Concateno los DataFrames
tienda_completa = pd.concat([tienda_1, tienda_2, tienda_3, tienda_4], ignore_index=True)
```

```python
# Agrupo por tienda y sumo la facturación
facturacion_por_tienda = tienda_completa.groupby('id_tienda')['Precio'].sum()
print("Facturación total por tienda:")
print(facturacion_por_tienda)
```

```python
import matplotlib.pyplot as plt

facturacion_por_tienda.plot(kind='bar', color='skyblue')
plt.title('Facturación Total por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Facturación Total')
plt.show()
```

#2. Ventas por categoría#


```python
import pandas as pd


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"


tienda_1 = pd.read_csv(url1)
tienda_2 = pd.read_csv(url2)
tienda_3 = pd.read_csv(url3)
tienda_4 = pd.read_csv(url4)


tienda_1['id_tienda'] = 'Tienda 1'
tienda_2['id_tienda'] = 'Tienda 2'
tienda_3['id_tienda'] = 'Tienda 3'
tienda_4['id_tienda'] = 'Tienda 4'


tienda_completa = pd.concat([tienda_1, tienda_2, tienda_3, tienda_4], ignore_index=True)


print(tienda_completa.head())
```

```python
# tienda y categoría y la cuenta de las ventas
ventas_por_categoria = tienda_completa.groupby(['id_tienda', 'Categoría del Producto']).size()

# DataFrame con sort_values, primero por tienda y luego por ventas dentro de cada tienda
ventas_por_categoria = ventas_por_categoria.sort_values(ascending=False).reset_index(name='Cantidad de Ventas')

# Ventas organizadas por tienda (de mayor a menor dentro de cada tienda)
print("Ventas por categoría en cada tienda (ordenadas de mayor a menor dentro de cada tienda):")
print(ventas_por_categoria)


```

```python
# Sumar ingresos por tienda y categoría
ventas_tienda_categoria = tienda_completa.groupby(['id_tienda', 'Categoría del Producto'])['Precio'].sum().unstack()

# Crear el gráfico de barras apiladas
ventas_tienda_categoria.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab10')
plt.title('Ventas por Categoría en Cada Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos Totales')
plt.legend(title='Categorías del Producto', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=0)
plt.show()
```

```python
# Total de ingresos por categoría
ganancias_por_categoria = tienda_completa.groupby('Categoría del Producto')['Precio'].sum()


print("Ingresos por categoría:")
print(ganancias_por_categoria)
```

```python
ganancias_por_categoria.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90)
plt.title('Distribución de Ingresos por Categoría')
plt.ylabel('')
plt.show()
```

# 3. Calificación promedio de la tienda#


```python
# Calculo promedio de calificaciones por tienda
calificacion_promedio_por_tienda = tienda_completa.groupby('id_tienda')['Calificación'].mean()
calificacion_porcentual = (calificacion_promedio_por_tienda / 5) * 100

print("Calificación promedio por tienda:")
print(calificacion_promedio_por_tienda)
```

```python
import matplotlib.pyplot as plt
calificacion_porcentual = calificacion_porcentual.astype(float)
ax = calificacion_porcentual.plot(kind='bar', color='lightgreen', figsize=(8, 6))
plt.title('Calificación Promedio en Porcentajes por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Calificación Promedio (%)')
plt.xticks(rotation=0)


for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}%", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

#4. Productos más y menos vendidos#
##frecuencia##

```python
# Contando cantidad de veces que se vendió algo
productos_vendidos = tienda_completa['Producto'].value_counts()


# Producto más vendido y el menos vendido
producto_mas_vendido = productos_vendidos.idxmax()
producto_menos_vendido = productos_vendidos.idxmin()

print(f"Producto más vendido: {producto_mas_vendido}, con {productos_vendidos.max()} ventas")
print(f"Producto menos vendido: {producto_menos_vendido}, con {productos_vendidos.min()} ventas")
```

##ingreso##


```python
# Suma de ingresos generados por cada producto
ingresos_por_producto = tienda_completa.groupby('Producto')['Precio'].sum()

# Producto más y menos vendido por ingresos
producto_mas_ingresos = ingresos_por_producto.idxmax()
producto_menos_ingresos = ingresos_por_producto.idxmin()

print(f"Producto con más ingresos: {producto_mas_ingresos}, con ${ingresos_por_producto.max():,.2f}")
print(f"Producto con menos ingresos: {producto_menos_ingresos}, con ${ingresos_por_producto.min():,.2f}")
```

```python
# 10 productos más vendidos (FRECUENCIA)
productos_top_10 = productos_vendidos.head(10)

import matplotlib.pyplot as plt

productos_top_10.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Top 10 Productos Más Vendidos')
plt.xlabel('Productos')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.show()
```

```python
# 10 productos con mayores ingresos
ingresos_top_10 = ingresos_por_producto.nlargest(10)

ingresos_top_10.plot(kind='bar', color='lightgreen', figsize=(10, 6))
plt.title('Top 10 Productos con Más Ingresos')
plt.xlabel('Productos')
plt.ylabel('Ingresos Totales')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
facturacion_total = [1800000, 1500000, 1600000, 1400000]  # Facturación total por tienda
calificacion_promedio = [79.53, 80.75, 80.97, 79.92]  # Calificación promedio (%)
costo_envio = [12500, 14500, 11200, 16000]  # Costo promedio de envío
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Normalizar datos (para hacerlos comparables en el gráfico)
facturacion_total_norm = [x / max(facturacion_total) * 100 for x in facturacion_total]
costo_envio_norm = [x / max(costo_envio) * 100 for x in costo_envio]

# Crear posiciones para las barras
x = np.arange(len(tiendas))

# Crear el gráfico de barras
width = 0.3  # Ancho de cada barra

plt.bar(x - width, facturacion_total_norm, width, label='Facturación Total (Normalizada)', color='skyblue')
plt.bar(x, calificacion_promedio, width, label='Calificación Promedio (%)', color='lightgreen')
plt.bar(x + width, costo_envio_norm, width, label='Costo de Envío (Normalizado)', color='orange')

# Agregar detalles
plt.title('Comparación de Rentabilidad por Tienda')
plt.xlabel('Tiendas')
plt.ylabel('Valores (Escala Normalizada)')
plt.xticks(x, tiendas)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
